```python
from random import randint
```

```python
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]
```

```python
#Instead of calling .plot, we can call .scatter to create a scatter chart. Unlike .plot, .scatter requires two 
#data series to plot against each other.
```

```python
import math # add dotted line, note: uses a separate plot (line) to get the line

x = range(-4,4)
y = [math.sin(n) for n in x]
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('An approximate distribution for sin(x)')
plt.plot(x,y, ls = ":") # dotted line
plt.show()
```

```python
x = range(-4,4)
y = [math.sin(n) for n in x]
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('An approximate distribution for sin(x)')
plt.plot(x,y, ls = "--") #with dash line 
plt.show()
```

```python
# We can use the .text to add arbitrary text to the chart.

# We'll need to specify x and y coordinates for where the text should be placed. The coordinates provided 
# specify where the left boundary of the text starts.

# We can also specify the color and fontsize keyword arguments to change the presentation of the text.
```

```python
x = list(range(-4,4))
x.append(3.14159)# to have a specific point on the line graph be shown
y = [math.sin(n) for n in x]

plt.scatter(x,y)
plt.plot(x,y, ls = '--')
plt.text(.25,0, "(0, 0)", fontsize = 9, c = 'red')
plt.text(1.6, .1, "($\pi$,sin($\pi$))", fontsize = 13, c = 'green')
plt.show()

```

```python
#Annotations can be even more fine-grained with the .annotate method.
```

```python
x = range(-3,10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x,y, c = 'orange')
plt.plot(x,z, c = 'blue')

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'green'})

plt.show()
```

```python
x = range(-6,5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize = (14,6)) #width, height

plt.scatter(x, sin_x, c = 'red')
plt.scatter(x, cos_x, c = 'purple')
plt.plot(x, sin_x, ls = ":", c = 'blue', alpha = .8, label = '$\sin(x)$')
plt.plot(x, cos_x, ls = "--", c = 'green', label = "$\cos(x)$")

plt.legend(loc = 'lower left')
plt.title('Approximate sine and cosine distributions', fontsize = 14)
plt.xlabel("$x$")

plt.show()
```

```python
# We can create multiple charts together using the .subplots function. We'll need to specify the total number of rows, 
# total number of columns, and which chart is currently the "active" chart.
```

```python
n_rows = 1
n_cols = 2


# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x ~ z')

plt.show()
```

```python
# Using subplots is very common, and the number of rows, columns, and active chart can be be specified 
# in a short hand of just one number.
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121) #rows, columns and chart number have been consolidated to 1 number
plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(122) #rows, columns and chart number have been consolidated to 1 number
plt.plot(x, z)
plt.title('x ~ z')
plt.suptitle('Examples of subplots') #supertitle over the individual subplots

plt.show()
```

```python
#The align property tells matplotlib how the bins should be aligned over the data that they represent (3 options. 
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.figure(figsize = (16,6))
plt.suptitle(f"x = {x}")

plt.subplot(131)
plt.hist(x, bins = [0,1,2,3,4,5,6])
plt.title('align = "middle" (default)')

plt.subplot(132)
plt.hist(x, bins = [0,1,2,3,4,5,6])
plt.title('align = "left" ')

plt.subplot(133)
plt.hist(x, bins = [0,1,2,3,4,5,6])
plt.title('align = "right"')

plt.show()
```

```python
plt.figure(figsize = (12,6))
for i in range(1,5):
    n = 10 ** i
    x1 = [randint(1,6) for _ in range(n)]
    x2 = [randint(1,6) for _ in range(n)]
    
    total = [x + y for x,y in zip(x1,x2)]
    plt.subplot(2,2,i)
    plt.hist(total, bins=range(2, 14), align='left', color='yellow', edgecolor='blue')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()
```

```python
# If matplotlib were limited to working with lists, it would be fairly useless for numeric processing. 
# Generally, you will use numpy arrays. In fact, all sequences are converted to numpy arrays internally. 
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
