```python
# Use matplotlib to plot the following equation:

# y = x ** 2 - x + 2

# Add an anotation for the point 0, 0, the origin.
```

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(-50, 50))
y = [(n ** 2 -n + 2) for n in x]
plt.scatter(x, y)
plt.title('A Quadratic Distribution, $y = x^2$')
plt.xlabel('$x$')
plt.ylabel('$x^2$')
plt.annotate('Origin', xy=(0, 0), xytext=(3, 1000),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14, 6)) # (width, height)
x = np.linspace(0, 50, 150)
y = [math.tan(n) for n in x]


plt.plot(x, y, c='blue', label='$\tan(x)$')
plt.xlabel('$x$')
plt.ylim(-10,10)
plt.show()
```

```python
x = range(0,50)
y = [2 ** n for n in x]

plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
x = np.arange(0, 25, .1)
y1 = [n ** .5 for n in x]
y2 = [n ** 3 for n in x]
y3 = [math.tan(n) for n in x]
y4 = [2 ** n for n in x]

plt.subplot(2, 2,1)
plt.plot(x,y1)
plt.subplot(2, 2,2)
plt.plot(x,y2)
plt.subplot(2, 2,3)
plt.plot(x,y3)
plt.subplot(2, 2,4)
plt.plot(x,y4)

plt.suptitle('Individual charts')
plt.show()
    
    
```

```python
plt.figure(figsize=(12, 8))
x = np.arange(0, 25, .1)
y1 = [n ** .5 for n in x]
y2 = [n ** 3 for n in x]
y3 = [math.tan(n) for n in x]
y4 = [2 ** n for n in x]


plt.plot(x,y1, c= 'blue', label = "$Sqrt$")
plt.plot(x,y2, c= 'green', label = "$Qube$")
plt.plot(x,y3, c = 'yellow', label = "$Tangent$")
plt.plot(x,y4, c = 'red', label = "$2 to the X$")

plt.legend(loc = "upper left")
plt.suptitle('Combined charts')
plt.xlabel("$x$")
plt.show()
```

```python
import math

x = np.arange(-4, 4, 1) #default scale is 1 (optional to include here)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls=':')
plt.title('An approximation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
