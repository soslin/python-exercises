```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# Merging is like a JOIN in SQL. The merge function will accept the left and right dataframes, and two keyword 
# arguments, left_on, and right_on that are used to indicate the columns that are keys that will be used to merge 
# the dataframes together. In addition, the how keyword argument is used to define what type of JOIN we want to 
# do; one of inner, left, right, or outer.
```

```python
#right join
```

```python
pd.merge(users, roles, left_on='role_id', right_on='id', how='right')
```

```python
#left join
```

```python
pd.merge(users, roles, left_on='role_id', right_on='id', how='right')
```

```python
#inner join
```

```python
pd.merge(users, roles, left_on='role_id', right_on='id', how='inner')
```

```python
inner_merged = pd.merge(users, roles, left_on='role_id', right_on='id', how='inner')
inner_merged
```

```python
inner_merged.drop('id_y', axis = 1)
```

```python
#rename columns while merging
```

```python
pd.merge(
    users.rename(columns = {'id' : 'user_id', 'name': 'user_name'}),
    roles.rename(columns = {'name' : 'role'}),
                left_on='role_id', right_on='id', how='inner')
```
